# Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X = car_data.drop(['Car_Name','Selling_Price'],axis=1)
y = car_data['Selling_Price']
```

```python
# Step 1: Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Step 4: Evaluate the model
y_pred = model.predict(X_test)
```

```python
# Calculate performance metrics
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (R2): {r2:.2f}")
```

```python
plt.figure(figsize=(10, 6))

# Plot actual selling prices in blue
plt.scatter(y_test, y_test, color='blue', label='Actual Selling Price')

# Plot predicted selling prices in red
plt.scatter(y_test, y_pred, color='red', alpha=0.5, label='Predicted Selling Price')

plt.xlabel('Actual Selling Price')
plt.ylabel('Predicted Selling Price')
plt.title('Actual vs. Predicted Selling Prices (Linear Regression)')
plt.legend()
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# Create a Random Forest Regressor model
rf_model = RandomForestRegressor(random_state=17)
```

```python
# Fit the model to the training data
rf_model.fit(X_train, y_train)
```

```python
# Make predictions on the test data
rf_y_pred = rf_model.predict(X_test)
```

```python
# Calculate performance metrics for the Random Forest model
rf_mse = mean_squared_error(y_test, rf_y_pred)
rf_r2 = r2_score(y_test, rf_y_pred)
```

```python
print(f"Random Forest Mean Squared Error (MSE): {rf_mse:.2f}")
print(f"Random Forest R-squared (R2): {rf_r2:.2f}")
```

```python
plt.figure(figsize=(10, 6))
# Plot actual selling prices in blue
plt.scatter(y_test, y_test, color='blue', label='Actual Selling Price')

# Plot predicted selling prices in red
plt.scatter(y_test, rf_y_pred, color='red', alpha=0.5, label='Predicted Selling Price')

plt.xlabel('Actual Selling Price')
plt.ylabel('Predicted Selling Price')
plt.title('Actual vs. Predicted Selling Prices (Random Forest)')
plt.legend()
plt.show()
```

```python
# Get feature importances from the trained model
feature_importances = rf_model.feature_importances_

# Create a DataFrame to display feature importances
feature_importance_df = pd.DataFrame({'Feature': X.columns, 'Importance': feature_importances})

# Sort the DataFrame by importance in descending order
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Plot the feature importances
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df, palette='viridis')
plt.title('Feature Importance')

plt.show()
```

Present price, Year, Driven Kms, Transmission, Selling Type, Fuel Type and Owner are the least important features for car price prediction.
